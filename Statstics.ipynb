```python
import numpy as np
import seaborn as sns
import pylab as py
import statsmodels.api as sm
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
```

```python
data = np.random.lognormal(0,1,100) # Here we have taken log noramally distributed data
df = pd.DataFrame(data)
```

```python
dist=sns.distplot(data,kde=True, hist=False)
plt.show()
```

```python
sm.qqplot(data, line='45')
# QQ plot of log normal distributed data
```

# Another E.G.

```python
original_data = np.random.exponential(size=1000)
# pareto distributed data

# Transform training data and save lambda value
fitted_data, fitted_lambda = stats.boxcox(original_data)


print(f'lambda value is : {fitted_lambda}')
```

```python
# plotting original data
sns.distplot(original_data, hist = False, kde=True,
            kde_kws = {'shade' : True, 'linewidth':2},
            label = 'Non-Normal',color='green')
plt.show()
```

```python
sm.qqplot(original_data,line='45')
# QQ plot of original data
```

```python
# plotting fitted data
sns.distplot(fitted_data, hist = False, kde=True,
            kde_kws = {'shade' : False, 'linewidth':2},
            label = 'Normal',color='red')
plt.show()
```

```python
sm.qqplot(fitted_data,line='45')
# qq plot of fitted data
```
